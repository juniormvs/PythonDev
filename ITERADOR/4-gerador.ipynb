Yield

É uma palavra chave para criar geradores. Qaundo você usa ele em uma função, você transforma a função em um gerador, que é um tipo especial de iterador.

Return: Quando é usado numa função, é encerrado e o valor é retornado. Chamadas subsequentes à função iniciam a execução novamente do código.

Yield: Pode ser chamada novamente e retormas a execução do ponto onde foi pausada, permitindo a produção de uma sequência de valores ao longo do tempo.

```python
def numeros_impares(n):
    for i in range(1, n+1, 2):#começa a contar do número 1,e pula de dois em dois até o valor passado pelo parametro n
        yield i
```

```python
for numero in numeros_impares(10):    
    print(numero)
```
