Enumerate
- Consegue associar o valor e também o índice de uma lista


```python
frutas = ["Maça", "Banana", "Laranja"]
```

```python
for indice, fruta in enumerate(frutas):
    print(f"{indice}-{fruta}")
```

```python
numeros = [10,15,20,25,30]
```

```python
indice_valores = {indice: valor for indice, valor in enumerate(numeros)}
indice_valores
```

```python
indice_valores.items()
```

```python
indices_pares = {indice: valor for indice, valor in indice_valores.items() if indice % 2 == 0}
indices_pares
```
