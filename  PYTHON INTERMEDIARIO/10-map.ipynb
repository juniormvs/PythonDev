Map
- Função que aplica uma função a todos os itens de um iterável (uma lista, por exemplo).
- É uma maneira eficiente e consisa de transformar dados.

```python
def quadrado(x):
    return x ** 2
```

```python
numeros = [1,2,3,4,5,6]
```

```python
for n in numeros:
    print(quadrado(n))
```

```python
resultados = map(quadrado, numeros)
resultados #Diz que é um map

list(resultados)#resultados já na lista
```

```python
def celsius_para_fahrenheit(c):
    return (c*9/5) + 32
```

```python
temperaturas_celsius = [10, 20, 25, 30, 35]
```

```python
temperaturas_fahreinheit = map(celsius_para_fahrenheit, temperaturas_celsius)
list(temperaturas_fahreinheit)
```
