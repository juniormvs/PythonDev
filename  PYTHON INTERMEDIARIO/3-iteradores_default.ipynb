```python
import itertools
```

```python
#1 - Exemplo Range()
for i in range(5):
    print(i)

```

```python
#2 - Iterar Listas
frutas =["maçã","laranja","abacaxi","banana","melão"]
for fruta in frutas:
    print(fruta)
```

```python
#3 - Iterar Dicionário
pessoa = {'Nome':'Alice','idade':33}
for chave in pessoa:
    print(chave, pessoa[chave])
```

```python
#4 - Enumerate()
for indice, fruta in enumerate(frutas):
    print(indice,fruta)
```

```python
#5 - Zip()
nomes = ['João', 'Ana', 'Marcelo', 'Beatriz']
idades = [20,30,29,22]

for nome, idade in zip(nomes, idades):
    print(nome, idade)
```

```python
#6 - Itertools
for i in itertools.count(start=10, step=2):#começa com 10 e pula de dois em dois números até chegar ao 20
    if i > 20:
        break
    print(i)
```
