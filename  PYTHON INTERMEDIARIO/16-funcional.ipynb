Programação Funcional
- Trata a programação como a avaliação de funções matemáticas e evita estados e dados mutáveis

```python
# Função de alta ordem - é uma função que recebe outra função como argumento
def quadrado(x):
    return x * x

def aplicar_funcao(funcao,lista):
    return[funcao(x) for x in lista]
```

```python
numeros = [1,2,3,4,5]
resultado = aplicar_funcao(quadrado, numeros)
resultado
```

```python
# Dados de Exemplo - dicionário
usuarios = [
    {'nome': 'Alice', 'idade' : 30, 'salario': 2000},
    {'nome': 'Bob', 'idade' : 25, 'salario': 3000},
    {'nome': 'Charlie', 'idade' : 35, 'salario': 5000},
    {'nome': 'Maria', 'idade' : 28, 'salario': 2200},
    {'nome': 'Ana', 'idade' : 22, 'salario': 2800},
]
```

```python
def extrair_nomes(usuarios):
    return list(map(lambda u: u['nome'], usuarios))

def extrair_nomes_e_salarios(usuarios, salario):
    return list(filter(lambda u: u['nome'] and u['salario'] > salario, usuarios))


def filtrar_maiores_que(usuarios, idade):
    return list(filter(lambda u: u['idade'] > idade, usuarios))

def calcular_media_idade(usuario):
    if not usuarios:
        return 0
    total_idade = sum(map(lambda u: u['idade'], usuarios))
    return total_idade / len(usuarios)
```

```python
usuarios_maiores_que_27 = filtrar_maiores_que(usuarios, 27)
usuarios_maiores_que_27
```

```python
nomes_maiores_que_27 = extrair_nomes(usuarios_maiores_que_27)
nomes_maiores_que_27
```

```python
idades_maiores_que_27 = filtrar_maiores_que(usuarios,27)
idades_maiores_que_27
```

```python
salarios_maiores_que_2000 = extrair_nomes_e_salarios(usuarios, 2000 )
salarios_maiores_que_2000
```

```python
media_idade = calcular_media_idade(usuarios_maiores_que_27)
media_idade
```

MÓDULO LRU

```python
from functools import lru_cache
```

```python
@lru_cache(maxsize=None) #Gerenciamento de memária
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n-2)
```

```python
for i in range(10):
    print(f"fibonacci({i}) = {fibonacci(i)}")
```
